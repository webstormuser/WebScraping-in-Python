```python
#Loading and importing libraries
from bs4 import BeautifulSoup as bs4 
import pandas as pd 
import numpy as np 
import requests #used to request the url for which we want to scrape the information 
import urllib
import re
from urllib.request import urlopen 
```

```python
#!pip install requests_html
baseurl='https://www.flipkart.com/'
```

```python
products_links=[]
for x in range (1,21):
    r=requests.get(f'https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.price_range.from%3D10000&p%5B%5D=facets.price_range.to%3D30000&page={x}')
    soup=bs4(r.content,'lxml')
    product_list=soup.find_all('div',class_='_13oc-S')
    for item in product_list:
        for link in item.find_all('a',href=True):
            products_links.append('https://www.flipkart.com'+link['href'])
print(len(products_links))
PRODUCTS=[]
for link in products_links:
    r=requests.get(link,headers={"User-Agent":"Mozilla/5.0"})
    html_page=bs4(r.content,'lxml')
    name=html_page.find('div',class_='aMaAEs').find('h1',class_='yhB1nd').text.strip()
        #print(name)
    price=html_page.find('div',class_='_30jeq3 _16Jk6d').text.strip()
        #print(price)
    try:
        offer=html_page.find('div',class_='_3Ay6Sb _31Dcoz').span.text
    except:
        offer='no offer'
    try:
        ratings=html_page.find('div',class_='_3LWZlK').text.strip()
    except:
        ratings='no ratings'
    votes=html_page.find('span',class_='_2_R_DZ').text.strip()
    specification=html_page.find_all('li',class_='_21Ahn-')
    highlights=[]
    for i in specification:
            highlights.append(i.text.strip())
           # print(highlights)
  
    #PRODUCTS.append(name,price,offer,ratings,votes,highlights)
product={
    'name':name,
    'price':price,
    'offer':offer,
    'ratings':ratings,
    'votes':votes,
    'specification':highlights
    }
print(product)
PRODUCTS.append(product)
print('saving:',product[name])
df=pd.DataFrame(PRODUCTS)

```
