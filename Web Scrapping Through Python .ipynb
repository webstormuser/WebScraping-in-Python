```python
import openpyxl #loading these accesed data from website to excel file.
#import openpyxl module to load that data to excel file 
```

```python
#create new excel file 
excel_file=openpyxl.Workbook()
print(excel_file.sheetnames)#default sheet name provided By Python
sheet=excel_file.active
sheet.title='Top Rated Movies'
print(excel_file.sheetnames) #Now sheet name has changed
    #giving heading to excel column
sheet.append(["Rank","Movie Name","Year of Release","Rating"])
try:
    source=requests.get('https://www.imdb.com/chart/top/') #it fech the exact webpage which are going to scrap the data
    source.raise_for_status() # it raises error if we entered wrong URL by mistake so that user get idea to enter correct URL.
    soup=BeautifulSoup(source.text,'html.parser')# By default html parser provided by BeautifulSoup to fetch the HTML tags and parse it .
    #print(soup)#prints the html source code
    movies=soup.find('tbody',class_="lister-list").find_all('tr')# find() searches the first match in tbody section with class name lister-class where movies details are available .Whenever we want to extract class detail we must have to specify class_.find_all() method finds all the <tr> tags where all movies detail with their name ,year,ratings are availble .In find_all() we specify only tr tag because it does not contain any class .
    #print(len(movies))
    #We have to retrieve all the tr tags and inside tr there are td tags also .For each movie all detail must have to retrive we have to loop the code.
    for movie in movies:
        #a-->represents the <a> available in td class and inside that there is text data to access that we have to use text().
        name=movie.find('td',class_="titleColumn").a.text 
        rank=movie.find('td',class_='titleColumn').get_text(strip=True).split('.')[0] #strip()[0] removes all new line ,whitespaces .,split () splits the string based on. And we get exact number like 1.,2.,3. like 
        #print(rank)
        year=movie.find('span',class_='secondaryInfo').text.strip("()")
        rating=movie.find('td',class_="ratingColumn imdbRating").strong.text
        #print(rating)
        print(rank,name,year,rating)
        sheet.append([rank,name,year,rating])
except Exception as e:
    print(e)
excel_file.save("IMDB Movies Rating.xlsx")
```
